```python
import pandas as pd

# Load your data
eng = pd.read_csv('clean_preprocessed_email_engagements.csv')

# Ensure column names are clean and lowercase for safety
eng.columns = eng.columns.str.strip()

# Mark bounced emails
eng['is_bounced'] = eng['raw_status'].str.lower() == 'bounced'

# Total emails sent per prospect
total_sent = eng.groupby('Email').size().reset_index(name='total_sent')

# Total bounced emails per prospect
total_bounced = eng[eng['is_bounced']].groupby('Email').size().reset_index(name='total_bounced')

# Merge and compute bounce rate per email
bounce_rate_df = total_sent.merge(total_bounced, on='Email', how='left')
bounce_rate_df['total_bounced'] = bounce_rate_df['total_bounced'].fillna(0)
bounce_rate_df['email_bounce_rate'] = bounce_rate_df['total_bounced'] / bounce_rate_df['total_sent']

# Final result
bounce_rate_df = bounce_rate_df[['Email', 'email_bounce_rate']]
print(bounce_rate_df)
```

```python
import pandas as pd

# Convert to datetime in main DataFrame
eng['first_email_sent_date'] = pd.to_datetime(eng['first_email_sent_date'], errors='coerce')
eng['final_status_time'] = pd.to_datetime(eng['final_status_time'], errors='coerce')

# Total emails sent per Email
total_sent = eng.groupby('Email').size().reset_index(name='total_sent')

# Aggregate first and last dates per Email
date_range = eng.groupby('Email').agg(
    first_date=('first_email_sent_date', 'min'),
    last_date=('final_status_time', 'max')
).reset_index()

# Convert to datetime
date_range['first_date'] = pd.to_datetime(date_range['first_date'], errors='coerce')
date_range['last_date'] = pd.to_datetime(date_range['last_date'], errors='coerce')

# Remove timezones if present
date_range['first_date'] = date_range['first_date'].dt.tz_localize(None)
date_range['last_date'] = date_range['last_date'].dt.tz_localize(None)

# Calculate total days targeted (inclusive)
date_range['total_days_targeted'] = (date_range['last_date'] - date_range['first_date']).dt.days + 1
date_range['total_days_targeted'] = date_range['total_days_targeted'].replace(0, 1)

# Merge with total emails sent
engagement_df = total_sent.merge(date_range, on='Email', how='left')

# Compute engagement density
engagement_df['engagement_density'] = engagement_df['total_sent'] / engagement_df['total_days_targeted']

# Final output
engagement_df = engagement_df[['Email', 'engagement_density']]
print(engagement_df)
```

```python
import pandas as pd

# Convert email_sent_et and status_time_clean to datetime
eng['email_sent_et'] = pd.to_datetime(eng['email_sent_et'], errors='coerce')
eng['status_time_clean'] = pd.to_datetime(eng['status_time_clean'], errors='coerce')

# Handle timezone for email_sent_et
if eng['email_sent_et'].dt.tz is None:
    eng['email_sent_et'] = eng['email_sent_et'].dt.tz_localize('US/Eastern', ambiguous='NaT', nonexistent='NaT')
else:
    eng['email_sent_et'] = eng['email_sent_et'].dt.tz_convert('US/Eastern')

# Handle timezone for status_time_clean (date when email was opened)
if eng['status_time_clean'].dt.tz is None:
    eng['status_time_clean'] = eng['status_time_clean'].dt.tz_localize('US/Eastern', ambiguous='NaT', nonexistent='NaT')
else:
    eng['status_time_clean'] = eng['status_time_clean'].dt.tz_convert('US/Eastern')

# Extract day of week from status_time_clean (when email was opened)
eng['opened_day_of_week'] = eng['status_time_clean'].dt.dayofweek

# Filter for opened emails on weekends (Saturday=5, Sunday=6)
opened_weekends = eng[
    (eng['status'].str.lower() == 'open') &
    (eng['opened_day_of_week'].isin([5, 6]))
]

# Group by Email and count total opened emails on weekends
opened_weekends_df = opened_weekends.groupby('Email').size().reset_index(name='total_opened_weekends')

# Get all unique emails to ensure everyone is included in output
all_emails = eng[['Email']].drop_duplicates()

# Merge with all emails, fill missing counts with 0
opened_weekends_df = all_emails.merge(opened_weekends_df, on='Email', how='left')
opened_weekends_df['total_opened_weekends'] = opened_weekends_df['total_opened_weekends'].fillna(0).astype(int)

print(opened_weekends_df)
```
